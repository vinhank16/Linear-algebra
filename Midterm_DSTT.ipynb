### Thể hiện ma trận bằng numpy

```python
M1 = np.array([ [9, 12], [23,30] ])
u = np.array([ 2, 1])
tichuM1 = u.dot(M1)
print(tichuM1)
```

```python
#Chuyển thành dạng ma trận
np.asmatrix(M1)
```

```python
mat1 = np.zeros([5,5])
mat1
```

```python
A = np.mat(np.random.random((2,2)))
#Ma trận nghịch đảo
print(A.I)
#Ma trận chuyển vị
print(A.T)
```

```python
#Ma trận tam giác dưới
M = np.array([[-1,3,2],[0,-2,1],[1,5,-2]])
M_lower = np.tril(M)
print(M_lower, end = "\n\n")

#Ma trận tam giác trên
M_upper = np.triu(M)
print(M_upper, end = "\n\n")
```

```python
#Vector đường chéo
v_diag = np.diag(M)
print (v_diag, end = "\n\n")

#Chuyển vector thành ma trận đường chéo
M_diag = np.diag(v_diag)
print (M_diag, end = "\n\n")
```

```python
#Tạo ma trận đơn vị có đường chéo chính là 1, các pt còn lại là 0
I = np.identity(2)
I
```

```python
A = np.matrix([[1,-1,0,0],[0,1,-1,0],[0,0,0,-1],[-1,0,0,1]])
b = np.matrix([[-160], [240], [-600], [520]])
A_nghichdao = np.linalg.inv(A)
X = A_nghichdao*b
X
```
